**Пункт 9** Получить датафрейм с названием города и его населением из таблицы city

```python
df_city_population = pd.read_sql(
"""
SELECT
c.name,
c.population
FROM city c
""",
'postgresql://netology:NetoSQL2019@84.201.153.170:19001/world-db'    
) 
df_city_population.head()
   
```

**Пункт 10** При помощи Python соединить наш файл и таблицу с населением города. Сгруппировать итоговый дф по странами рассчитать сред.численность населения в городах отпуска. 

```python
df_merged = df_relax.merge(df_city_population, left_on= 'Название города на англ.языке', right_on='name', how='inner')
```

```python
df_merged
```

```python
df_last = df_merged.groupby('Страна', as_index=False).mean('population') #as_index=False сделали для того,чтобы столбец, по которому группировали остался столбцом, а не индексом строк
```
